```python
import pandas as pd
```

```python
# 파일들로 부터 데이터를 읽어온다.
filePath1 = 'https://raw.githubusercontent.com/bluehyena/Machine-Learning-Yahac/main/Chapter2/csv/lemonade.csv'

lemonade = pd.read_csv(filePath1)
lemonade
```

```python
filePath2 = 'https://raw.githubusercontent.com/bluehyena/Machine-Learning-Yahac/main/Chapter2/csv/boston.csv'

boston = pd.read_csv(filePath2)
boston
```

```python
filePath3 = 'https://raw.githubusercontent.com/bluehyena/Machine-Learning-Yahac/main/Chapter2/csv/iris.csv'

iris = pd.read_csv(filePath3)
iris
```

```python
# Dataset 으로 확인하기

print(lemonade.shape)
print(boston.shape)
print(iris.shape)
```

```python
# Col Name 출력

print(lemonade.columns)
print(boston.columns)
print(iris.columns)
```

```python
lem_indep = lemonade[['온도']]
lem_dep = lemonade[['판매량']]
print(lem_indep.shape, lem_dep.shape)

bos_indep = boston[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax',
       'ptratio', 'b', 'lstat']]
bos_dep = boston[['medv']]
print(bos_indep.shape, bos_dep.shape)
```
